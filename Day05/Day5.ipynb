## Supply Stacks

The expedition can depart as soon as the final supplies have been unloaded from the ships. Supplies are stored in stacks of marked crates, but because the needed supplies are buried under many other crates, the crates need to be rearranged.

The ship has a giant cargo crane capable of moving crates between stacks. To ensure none of the crates get crushed or fall over, the crane operator will rearrange them in a series of carefully-planned steps. After the crates are rearranged, the desired crates will be at the top of each stack.

The Elves don't want to interrupt the crane operator during this delicate procedure, but they forgot to ask her which crate will end up where, and they want to be ready to unload them as soon as possible so they can embark.

They do, however, have a drawing of the starting stacks of crates and the rearrangement procedure (your puzzle input). For example:
```
    [D]    
[N] [C]    
[Z] [M] [P]
 1   2   3 
```
- move 1 from 2 to 1
- move 3 from 1 to 3
- move 2 from 2 to 1
- move 1 from 1 to 2

In this example, there are three stacks of crates. Stack 1 contains two crates: crate Z is on the bottom, and crate N is on top. Stack 2 contains three crates; from bottom to top, they are crates M, C, and D. Finally, stack 3 contains a single crate, P.

Then, the rearrangement procedure is given. In each step of the procedure, a quantity of crates is moved from one stack to a different stack. In the first step of the above rearrangement procedure, one crate is moved from stack 2 to stack 1, resulting in this configuration:
```
[D]        
[N] [C]    
[Z] [M] [P]
 1   2   3 
```
In the second step, three crates are moved from stack 1 to stack 3. Crates are moved one at a time, so the first crate to be moved (D) ends up below the second and third crates:
```
        [Z]
        [N]
    [C] [D]
    [M] [P]
 1   2   3
```
Then, both crates are moved from stack 2 to stack 1. Again, because crates are moved one at a time, crate C ends up below crate M:
```
        [Z]
        [N]
[M]     [D]
[C]     [P]
 1   2   3
```
Finally, one crate is moved from stack 1 to stack 2:
```
        [Z]
        [N]
        [D]
[C] [M] [P]
 1   2   3
```
The Elves just need to know which crate will end up on top of each stack; in this example, the top crates are C in stack 1, M in stack 2, and Z in stack 3, so you should combine these together and give the Elves the message CMZ.

**After the rearrangement procedure completes, what crate ends up on top of each stack?**

## Part 1

### Read the stacks contents and build the models

````python
with open("crates.txt", "r") as f:
    stacks_text = f.read(36 * 9)
````

````python
# Stacks contents as read from file
stacks_text
````

````python
# List of stacks, as read from file, reversed
stacks_list = stacks_text.split("\n")[:-1]
stacks_list.reverse()
stacks_list
````

````python
# Stacks Dictionary
stacks_dict = {i: "" for i in range(1, 10)}
stacks_dict
````

````python
# Function to add a crate level to the stacks
def add_crate_level(dictionary, stacklist, level):
    # take one level of crates
    the_level = stacklist[level]
    # obtain just the crate IDs
    crateids = ""
    for i in range(1, len(the_level), 4):
        crateids += the_level[i]
    # iterate the level and add crates to stacks
    for i in range(0, len(crateids)):
        if crateids[i] != " ":
            dictionary[i+1] += crateids[i]
            print(f"Level {level}: added [{crateids[i]}] to stack {i+1}")
````

#### Create the initial working stacks dictionary

````python
for i in range(1, 9):
    add_crate_level(stacks_dict, stacks_list, i)
````

````python
stacks_list.reverse()
stacks_list
````

````python
stacks_dict
````

### Read the rearrangement procedures

````python
with open("crates.txt", "r") as f:
    procedure_raw = f.readlines()[10:]
````

````python
procedure_raw[:5]
````

````python
# [number_of_crates, stack_from, stack_to]
procedure_list = [[int(j[1]), int(j[3]), int(j[5])] for j in [i.rstrip("\n").split(" ") for i in procedure_raw]]
procedure_list[:10]
````

### Define the manouver function

````python
stacks_dict
````

````python
def manouver(stacks, action):
    # definitions
    crates = action[0]
    stack_from = action[1]
    stack_to = action[2]
    source = stacks[stack_from]
    destination = stacks[stack_to]
    moved = source[-crates:][::-1]
    print(f"{action}:  {stack_from}: {source} {source[:-crates]} | {stack_to}: {destination} {destination+moved}")    
    # do the move
    stacks[stack_from] = stacks[stack_from][:-crates]
    stacks[stack_to] += moved
````

### Operate the crane

````python
for action in procedure_list:
    manouver(stacks_dict, action)
````

### Obtain the top crate from all stacks

````python
stacks_dict
````

````python
for i in range(1, 10):
    print(stacks_dict[i][-1:], end="")
````
