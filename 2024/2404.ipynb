````python
def count_cross_mas(matrix):
    count = 0
    shape = len(matrix), len(matrix[0])

    # all cross_mas cases
    cross_mas_cases = [["M S", " A ", "M S"], ["S M", " A ", "S M"], ["M M", " A ", "S S"], ["S S", " A ", "M M"]]

    # Iterate
    for line in range(shape[0]-2):
        for column in range(shape[1]-2):
            # take the subset
            subset = [matrix[line][column:column+3], matrix[line+1][column:column+3], matrix[line+2][column:column+3]]
            # replace characters
            new_subset = [f"{subset[0][0]} {subset[0][2]}", f" {subset[1][1]} ", f"{subset[2][0]} {subset[2][2]}"]

            # add 1 for each case match
            for case in cross_mas_cases:
                count += 1 * (new_subset == case)

    return count
````
