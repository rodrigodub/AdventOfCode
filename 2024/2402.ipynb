--- Day 2: Red-Nosed Reports ---
Fortunately, the first location The Historians want to search isn't a long walk from the Chief Historian's office.

While the Red-Nosed Reindeer nuclear fusion/fission plant appears to contain no sign of the Chief Historian, the engineers there run up to you as soon as they see you. Apparently, they still talk about the time Rudolph was saved through molecular synthesis from a single electron.

They're quick to add that - since you're already here - they'd really appreciate your help analyzing some unusual data from the Red-Nosed reactor. You turn to check if The Historians are waiting for you, but they seem to have already divided into groups that are currently searching every corner of the facility. You offer to help with the unusual data.

The unusual data (your puzzle input) consists of many reports, one report per line. Each report is a list of numbers called levels that are separated by spaces. For example:

7 6 4 2 1
1 2 7 8 9
9 7 6 2 1
1 3 2 4 5
8 6 4 4 1
1 3 6 7 9
This example data contains six reports each containing five levels.

The engineers are trying to figure out which reports are safe. The Red-Nosed reactor safety systems can only tolerate levels that are either gradually increasing or gradually decreasing. So, a report only counts as safe if both of the following are true:

- The levels are either all increasing or all decreasing.
- Any two adjacent levels differ by at least one and at most three.

In the example above, the reports can be found safe or unsafe by checking those rules:

7 6 4 2 1: Safe because the levels are all decreasing by 1 or 2.
1 2 7 8 9: Unsafe because 2 7 is an increase of 5.
9 7 6 2 1: Unsafe because 6 2 is a decrease of 4.
1 3 2 4 5: Unsafe because 1 3 is increasing but 3 2 is decreasing.
8 6 4 4 1: Unsafe because 4 4 is neither an increase or a decrease.
1 3 6 7 9: Safe because the levels are all increasing by 1, 2, or 3.
So, in this example, 2 reports are safe.

Analyze the unusual data from the engineers. How many reports are safe?

````python
# Read file contents
with open("day-2-input.txt") as f:
    reports = [line.replace("\n", "") for line in f.readlines()]

# View data
reports
````

````python
def find_differences(report: str) -> list:
    # read all levels in a report
    levels = [int(level) for level in report.split(" ")]
    # find the differences between each level
    differences_list = [levels[index+1] - levels[index] for index in range(len(levels)-1)]
    return differences_list
````

````python
def test_trend(differences_list):
    # count signed values
    positives_count = len([i for i in differences_list if i >= 0])
    negatives_count = len([i for i in differences_list if i < 0])
    # return True if trend is constant
    # if positives_count == 0 or negatives_count == 0 and positives_count != negatives_count:
    #     return "Safe"
    # else:
    #     return "Unsafe"
    return positives_count == 0 or negatives_count == 0 and positives_count != negatives_count
````

````python
def test_adjency(differences_list):
    not_adjacents = [i for i in differences_list if (abs(i) > 3 or abs(i) < 1)]
    # return "Unsafe" if len(not_adjacents) > 0 else "Safe"
    return False if len(not_adjacents) > 0 else True
````

````python
red_nosed_reports = [[i, find_differences(i), test_trend(find_differences(i)), test_adjency(find_differences(i))] for i in reports]
red_nosed_reports
````

````python
safe_reports_count = len([i for i in red_nosed_reports if i[2] and i[3]])
safe_reports_count
# first try: 248 (too high)
````

````python
def dampen_report(report: str):
    # create a list from a report string
    a_report = report.split(" ")
    # output list
    # output = []

    # for each level in a report, remove it and ...
    for item in range(len(a_report)):
        removed_report = a_report.copy()
        removed_report.pop(item)
        # print(a_report[item], removed_report)
        # ... check if the shortened report is safe
        removed_report = " ".join(removed_report)
        removed_report_differences = find_differences(removed_report)
        removed_report_trend = test_trend(removed_report_differences)
        removed_report_adjency = test_adjency(removed_report_differences)

        # test if both tests are true
        if removed_report_trend and removed_report_adjency:
            # output.append([report, a_report[item], removed_report_trend, removed_report_adjency])
            return report, a_report[item], removed_report_trend, removed_report_adjency
        
    # Unchanged
    # if len(output) == 1:
    #     return output[0]
    # else:
    #     return None
    return None
````

````python
dampened_reports = [dampen_report(i) for i in unsafe_reports if dampen_report(i)]
dampened_reports
````
